```python
import pandas as pd
import numpy as np
```

```python
# Read in file 1
DF1 = pd.read_csv("student_scores.csv")
```

```python
DF1.head()
```

```python
# Read in file 2
DF2 = pd.read_csv("English_grades.csv")
```

```python
DF2.head()
```

```python
# Drop irrelevant column
DF1 = DF1.drop('Unnamed: 0', axis=1)
DF1.head()
```

```python
# merge DF1 and DF2
DF1 = DF1.merge(DF2, on='Name')
DF1.head()
```

```python
DF1 = DF1.drop('Unnamed: 0', axis=1)
DF1.head()
```

```python
DF1.info()
```

```python
# Get average age to fill in NA
avgAge= DF1['Age'].mean()
```

```python
DF1['Age'] = DF1['Age'].fillna(avgAge)
```

```python
DF1.info()
```

```python
# Fill in NA math scores with 0
DF1["Math Score"] = DF1["Math Score"].fillna(0)
DF1.info()
```

```python
DF1['City'].unique()
```

```python
# Replace 'Port' with 'Pt.'
DF1['City'] = DF1['City'].str.replace('Port', 'Pt.')
```

```python
DF1['City'].unique()
```

```python
DF1['Gender'] = DF1['Gender'].replace({'Male':'M', 'Female':'F', 'Other':'O'})
```

```python
DF1['Gender'].value_counts()
```

```python
#grouping function
DF1.groupby('Gender')['Math Score'].agg(['sum','mean'])
```

```python
#pivot table
table = pd.pivot_table(DF1, values = ['Math Score', 'Science Score', 'English Score'], index='Gender',
               aggfunc=np.mean, margins=True)
table
```

```python
table.info()
```

```python
df = DF1
```

```python
#Transform
df['Total Score'] = df['Math Score'] + df['English Score'] + df['Science Score']
df['Average Score'] = df.groupby('Gender')['Total Score'].transform('mean')
```

```python
df.head()
```

```python
# numpy conditional
avg_math = df['Math Score'].mean()
df['Above Average Math Score'] = np.where(df['Math Score']>avg_math, 'Yes', 'No')
# df.loc[df['Above Average Math Score']=='Yes']
# df.where(df['Above Average Math Score']=='Yes')
```

```python
# Create function
def AboveAvg(dataset, col, new_col):
    avg_value = dataset[col].mean()
    dataset[new_col] = np.where(dataset[col]>avg_value, 'Yes', 'No')
    return dataset

AboveAvg(df, 'English Score', 'Above Average English Score')
AboveAvg(df, 'Science Score', 'Above Average Science Score')
```

```python
df.to_csv('final_output_20240903.csv')
```
