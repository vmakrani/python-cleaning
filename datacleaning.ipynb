```python
import pandas as pd
import numpy as np

'''
This is a project where I imported two separate csv files into dataframes and then merged them on a common column. 
I also performed several functions in order to clean up nulls and other data inconsistencies. 
Once my dataset was clean, I performed several data analysis tasks such as calculating different levels of aggregations 
using group by, transform, and pivot_table.
'''
```

```python
# Read in file 1
DF1 = pd.read_csv("student_scores.csv")
```

```python
# Display first 5 records
DF1.head()
```

```python
# Read in file 2
DF2 = pd.read_csv("English_grades.csv")
```

```python
# Display first 5 records
DF2.head()
```

```python
# Drop irrelevant column from dataset 1
DF1 = DF1.drop('Unnamed: 0', axis=1)
DF1.head()
```

```python
# merge DF1 and DF2 on common column
DF1 = DF1.merge(DF2, on='Name')
DF1.head()
```

```python
# Drop irrelevant column from merged dataset
DF1 = DF1.drop('Unnamed: 0', axis=1)
DF1.head()
```

```python
# Info function gives us number of values and their data types
DF1.info()
```

```python
# Get average age to fill replace NA values for age column
avgAge= DF1['Age'].mean()
```

```python
DF1['Age'] = DF1['Age'].fillna(avgAge)
```

```python
# Check to see if Age now has 110 values
DF1.info()
```

```python
# Fill in NA math scores with 0
DF1["Math Score"] = DF1["Math Score"].fillna(0)
DF1.info()
```

```python
# Take a look at all unique values of City
DF1['City'].unique()
```

```python
# Replace 'Port' with 'Pt.'
DF1['City'] = DF1['City'].str.replace('Port', 'Pt.')
```

```python
# Check to see if string replace worked correctly
DF1['City'].unique()
```

```python
# Replace Gender values with letters only
DF1['Gender'] = DF1['Gender'].replace({'Male':'M', 'Female':'F', 'Other':'O'})
```

```python
DF1['Gender'].value_counts()
```

```python
# do a grouping function to find sum and mean of math scores
DF1.groupby('Gender')['Math Score'].agg(['sum','mean'])
```

```python
# pivot table to find the mean scores by Gender
# save this information to a new dataframe
table = pd.pivot_table(DF1, values = ['Math Score', 'Science Score', 'English Score'], index='Gender',
               aggfunc=np.mean, margins=True)
table
```

```python
table.info()
```

```python
# perform a Transform to insert Average Score into dataframe by Gender
DF1['Total Score'] = DF1['Math Score'] + DF1['English Score'] + DF1['Science Score']
DF1['Average Score'] = DF1.groupby('Gender')['Total Score'].transform('mean')
```

```python
DF1.head()
```

```python
# numpy conditional statement - which records are greater than the average math score?
avg_math = DF1['Math Score'].mean()
DF1['Above Average Math Score'] = np.where(DF1['Math Score']>avg_math, 'Yes', 'No')
# Below rows can be used to query dataframe for 'yes' values
DF1.loc[DF1['Above Average Math Score']=='Yes']
# DF1.where(DF1['Above Average Math Score']=='Yes')
```

```python
# Create function to calculate 'AboveAvg' for other columns
def AboveAvg(dataset, col, new_col):
    avg_value = dataset[col].mean()
    dataset[new_col] = np.where(dataset[col]>avg_value, 'Yes', 'No')
    return dataset

# Call Function to populate dataframe
AboveAvg(DF1, 'English Score', 'Above Average English Score')
AboveAvg(DF1, 'Science Score', 'Above Average Science Score')
```

```python
# write final output to a file
DF1.to_csv('final_output_20240903.csv')
```
